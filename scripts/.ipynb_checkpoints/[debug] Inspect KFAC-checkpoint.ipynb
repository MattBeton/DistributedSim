```python
import torch
import os

import matplotlib.pyplot as plt
```

```python
factors = [(x, torch.load(x)) for x in os.listdir() if 'pt' in x]
```

```python
for path, factor in factors:
    if not (('final' in path) or ('rhs' in path) or ('theta' in path)):
        print(path)
        plt.title(path)
        vmax = factor.abs().max().item() * 0.1
        plt.imshow(factor.detach().cpu(), vmin=-vmax, vmax=vmax)
        plt.xlabel(f'absmax: {vmax}')
        plt.show()
```
